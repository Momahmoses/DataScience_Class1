```python
import numpy as np
import pandas as pd
df = pd.read_csv('Big_Black_Money_Dataset.csv')
df.head()
```

```python
#

# 1. What percentage of missing or null values exist in each column? 
# 2.How would you handle these missing values/

missing_value_percentage = (df.isnull().sum() / len(df) * 100)

total_missing = df.isnull().sum()
missing_values_outcome= pd.concat([total_missing, missing_value_percentage],
                                  axis = 1)
missing_values_outcome
```

```python
# 2. PERFORM A SUMMARY OF BASIC STATISTICS ( MEAN, MEDIAN, STANDARD DEVIATION, MIN, AND MAX)
# FOR THE AMOUNT (USD) COLUMN. IDENTIFY ANY EXTREME OUTLIERS USING Z-SCORES OR IQR METHODS
df['Amount (USD)'].describe()
```

```python
# identify any extreme outliers using Z-scored:
from scipy import stats
z_scores = stats.zscore(df['Amount (USD)'])
outliers = df[(z_scores > 3) | (z_scores < -3)]
print(outliers)
```

```python
# USING THE INTER QUARTILE ERASE METHOD

Q1 = df['Amount (USD)'].quantile(0.25)
Q3 = df['Amount (USD)'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_r = df[(df['Amount (USD)'] < lower_bound) | (df['Amount (USD)'] > upper_bound)]
print(outliers_r)
```

```python
# 3. HOW MANY UNIQUE VALUES ARE PRESENT IN THE COUNTRY AND DESTINATION COUNTRY COLUMNS? 
# ARE THERE ANT DESCREPANCIES (E.G SPELLING INCONSISTENCIES)

col= df[['Country', 'Destination Country']]
for x in col:
    print(df[x].unique())
```

```python
# 4. convert the date of transaction column into components like year, month and day. WHAT are the most frequent transaction years and month


df.info()
```

```python
df['Date of Transaction'] = pd.to_datetime(df['Date of Transaction'])
df['Year_ of_transaction'] = pd.DatetimeIndex(df['Date of Transaction']).year
df['Month_ of_transaction'] = pd.DatetimeIndex(df['Date of Transaction']).month
df['Date_ of_transaction'] = pd.DatetimeIndex(df['Date of Transaction']).day
df.head()
```

```python
df['Year_ of_Transaction'].value_counts().sort_values(ascending = False)

df['Year_ of_Transaction'].value_counts().sort_values(ascending = False)
```

```python

# Perform frequency counts of categorical columns such as Transacetion Type and industry.
# Are there any values that occur unusually often or rarely 

transaction_type_counts = df['Transaction Type'].value_counts()
industry_counts = df['Industry'].value_counts()
print(transaction_type_counts)
print(industry_counts)
```

```python
# Are there any values that occur unusually often or rarely?
rarely_threshold = 0.02 * len(df)
often_threshold = 0.18 * len(df)
print(transaction_type_counts[transaction_type_counts < rarely_threshold])
print(industry_counts[industry_counts > rarely threshold])

print(transaction_type_counts[transaction_type_counts < often_threshold])
print(industry_counts[industry_counts > rarely threshold])
```
